```python
def applymode(x):
    a = x.mode()[0]
    print(a)
    return x.fillna(a)
```

```python
dataf["Occupation"]
```

```python
dataf["Occupation"]=dataf.groupby("City_Category").Occupation.transform(applymode)
```

```python
dataf.loc[dataf.Occupation ==0,"Occupation"]=np.nan
```

```python
dataf.isna().sum()
```

```python
dataf["Product_ID"].value_counts()
```

```python
dataf["Product_Id_Pattern"]=dataf['Product_ID'].apply(lambda x: x[0:4])
```

```python
dataf["Product_Id_Pattern"].value_counts()
```

```python
pandas_profiling.ProfileReport(dataf)
```

```python
def applymean(x):
    a = x.mean()
    print(a)
    return x.fillna(a)
```

```python
dataf["Product_Category_2"]=dataf.groupby("Product_Id_Pattern").Product_Category_2.transform(applymean)
```

```python
dataf["Product_Category_3"]=dataf.groupby("Product_Id_Pattern").Product_Category_3.transform(applymean)
```

```python
dataf.isna().sum()
```

```python
def value_all(x):
    print(x.dtype)
    if x.dtype=="object":

        print(x.value_counts())
#dataf.apply(lambda x :print(x.value_counts()))
dataf.apply(lambda x : value_all(x))
```

```python
data_column_category = dataf.select_dtypes(exclude=[np.number]).columns
```

```python
data_column_Integer = dataf.columns.difference(data_column_category)
```

```python
data_model = pd.concat([dataf_one_hot,dataf["Product_ID"],dataf[data_column_Integer]],axis=1)
```
